```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
path = ("/home/michael/DataSciMSc/comp1680/coursework_open_mp/graphs/jacobi/serial")
```

```python
def jacobiSingleGraph(x_array, y_array, xlabel, title, path, name):
    # plot with various axes scales
    plt.figure()
    fig, ax = plt.subplots()
    plt.xlabel(xlabel)
    plt.ylabel('Time (Seconds)')
    plt.title(title)
    plt.grid(True)
    ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
    plt.plot(x_array, y_array)
    plt.savefig("%s/%s.png"% (path,name))
    plt.show
```

```python
def jacobiCombinedGraph(x_label, Title, m, std, fst, o1, o2, o3, os, path, title):
    # plot with various axes scales
    plt.figure()
    fig, ax = plt.subplots()
    ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
    plt.xlabel(x_label)
    plt.yscale('linear')
    plt.ylabel('Time (Seconds)')
    plt.title(Title)
    plt.grid(True)
    plt.plot(m,std, label='Standard')
    plt.plot(m,fst,label='Fast')
    plt.plot(m,o1,label='O1')
    plt.plot(m,o2,label='O2')
    plt.plot(m,o3,label='O3')
    plt.plot(m,os,label='Small')
    plt.legend()
    plt.savefig("%s/%s.png"% (path,std))
```
